```python
x=df_new[df_new.columns[0:21]]
y=df_new['age_group']
x
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
scale.fit(x)
x=scale.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.metrics import f1_score
from sklearn import metrics
from sklearn import tree
age_tree=tree.DecisionTreeClassifier(criterion='entropy',max_depth=10)
age_tree.fit(x_train,y_train)
predtree=age_tree.predict(x_test)
print(f1_score(y_test,predtree,average='weighted'))
print(age_tree.score(x_train,y_train))
```

```python
from sklearn.metrics import f1_score
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(C=1,solver='liblinear').fit(x_train,y_train)
yhat2=lr.predict(x_test)
yhat_prob=lr.predict_proba(x_test)
print(f1_score(y_test,yhat2,average='weighted'))
print(lr.score(x_train,y_train))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=20)
rf.fit(x_train,y_train)
print(rf.score(x_test,y_test))
print(rf.score(x_train,y_train))
```

```python
from sklearn.ensemble import BaggingClassifier
bg=BaggingClassifier(tree.DecisionTreeClassifier(),max_samples=0.5,max_features=1,n_estimators=20)
bg.fit(x_train,y_train)
print(bg.score(x_test,y_test))
print(bg.score(x_train,y_train))
```

```python
from sklearn.ensemble import AdaBoostClassifier
adb=AdaBoostClassifier(tree.DecisionTreeClassifier(),n_estimators=10,learning_rate=1)
adb.fit(x_train,y_train)
print(adb.score(x_test,y_test))
print(adb.score(x_train,y_train))
```

```python
import lightgbm as lgb
```

```python
d_train=lgb.Dataset(x, label=y)
#setting up the parameters
params={}
params['learning_rate']=0.03
params['boosting_type']='gbdt' #GradientBoostingDecisionTree
params['objective']='multiclass' #Multi-class target feature
params['metric']='multi_logloss' #metric for multi-class
params['num_leaves']=50
params['max_depth']=10
params['num_class']=5 #no.of unique values in the target class not inclusive of the end value
#training the model
clf=lgb.train(params,d_train,100)
```

```python
#prediction on the test dataset
y_pred_1=clf.predict(x_test)
```

```python
#argmax() method 
y_pred_1 = [np.argmax(line) for line in y_pred_1]
y_pred_1
```
